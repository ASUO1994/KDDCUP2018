```python
import pandas
import numpy as np
import matplotlib.pyplot as plt



def dataset(mode): #mode取weather或air
    root = '/home/liyouru/.ipython/profile_myserver/dataset/KDD_Cup2018/'+mode+'.csv'
    data = pandas.read_csv(root)
    return data


mode = 'beijing_201802_201803_aq' #air day Figure
air = dataset(mode)
# print(air.describe())

gap_air = 24

PM10_miss = []
PM10_inter = []
index = 0
count = 0
for value in air['PM10']:

    if np.isnan(value):
        count += 1
        
    elif count != 0:
        PM10_miss.append(count)
        
        if count >= gap_air:
            PM10_inter.append(index - count)
            PM10_inter.append(index)
            
        count = 0
        
    index += 1

PM25_miss = []
PM25_inter = []
index = 0
count = 0
for value in air['PM2.5']:

    if np.isnan(value):
        count += 1
        
    elif count != 0:
        PM25_miss.append(count)
        
        if count >= gap_air:
            PM25_inter.append(index - count)
            PM25_inter.append(index)
            
        count = 0
        
    index += 1
    
NO2_miss = []
NO2_inter = []
index = 0
count = 0
for value in air['NO2']:

    if np.isnan(value):
        count += 1
        
    elif count != 0:
        NO2_miss.append(count)
        
        if count >= gap_air:
            NO2_inter.append(index - count)
            NO2_inter.append(index)
            
        count = 0
        
    index += 1

CO_miss = []
CO_inter = []
index = 0
count = 0
for value in air['CO']:

    if np.isnan(value):
        count += 1
        
    elif count != 0:
        CO_miss.append(count)
        
        if count >= gap_air:
            CO_inter.append(index - count)
            CO_inter.append(index)
            
        count = 0
        
    index += 1
    
O3_miss = []
O3_inter = []
index = 0
count = 0
for value in air['O3']:

    if np.isnan(value):
        count += 1
        
    elif count != 0:
        O3_miss.append(count)
        
        if count >= gap_air:
            O3_inter.append(index - count)
            O3_inter.append(index)
            
        count = 0
        
    index += 1
    
SO2_miss = []
SO2_inter = []
index = 0
count = 0
for value in air['SO2']:

    if np.isnan(value):
        count += 1
        
    elif count != 0:
        SO2_miss.append(count)
        
        if count >= gap_air:
            SO2_inter.append(index - count)
            SO2_inter.append(index)
            
        count = 0
        
    index += 1
    
PM10_inter.append(48676)
PM25_inter.append(48676)
NO2_inter.append(48676)
CO_inter.append(48676)
O3_inter.append(48676)
SO2_inter.append(48676)
print(max(PM10_miss))
print(PM10_inter)
```
