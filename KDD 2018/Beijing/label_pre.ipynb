```python
######sMAPE###########
def sMAPE(x, y):   
    s = 0
    print(x.shape)
    for i in range(x.shape[0]):
        s += np.abs(x[i] - y[i]) / (np.abs(x[i]) + np.abs(y[i])) / 2.
    print(s)
    s /= x.shape[0]
    return s
#     return np.mean(np.abs(x - y) / (np.abs(x) + np.abs(y)) / 2.)

#######rmse#################
def RMSE(x, y):
    s = 0
    for i in range(x.shape[0]):
        s += np.square(x[i] - y[i])
    print(s)
    return np.sqrt(s/len(y))
```
